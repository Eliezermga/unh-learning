```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data  = load_diabetes()
X_data, target = data.data , data.target
```

```python
## affichage des caracteristiques
print(data.DESCR)
```

```python
X_data[0]
```

## Regression lineaire
$$
\hat{y} = f(x_1, x_2, ... , x_n) = \sum_{i=1}^n \alpha_i x_i + \beta + \epsilon 
$$
- $x_i$: sont des variables independantes
- $y$ : c'est la variable dependante
- $\alpha_i$: sont des coefficents de regression
- $\beta$ : c'est le terme independant
- $\epsilon$ :  erreur aleatoire

## Erreur quadratique
$$
\text{erreur quadratique} = \sum_{j=1}^N(y_j-\hat{y}_j)^2
$$
##coefficent de determination $R^2$
$$
R^2 = 1-\frac{\sum_{j=1}^N {y_j-\hat{y}_j^2}{(y_j-\bar{y})^2}
$$
- $\bar{y}$: represente la valeur moyenne de la variable
- $N$ : le nombre d'individu repris dans le data set


```python
# split des donnees
X_train,X_test, y_train,y_test = train_test_split(X_data, target, test_size=0.3)
```

```python
regresseur = LinearRegression()
regresseur.fit(X_train, y_train)
```

```python
#test
y_pred= regresseur.predict(X_test)
```

```python
#netriques de performance
r2_score(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
plt.figure(figsize=(2,2))
plt.scatter(y_test, y_pred)
plt.xlabel("cible")
plt.ylabel("prediction")
plt.show()
```
